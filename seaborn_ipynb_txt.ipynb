```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

```python
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic)
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips);
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

```python
total_bill_thur = tips[tips['day']=='Thur']['total_bill']
total_bill_thur.describe()
```

```python
Q1 = total_bill_thur.quantile(0.25)
Q3 = total_bill_thur.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
total_bill_thur[total_bill_thur < (Q1 - 1.5 * IQR)]
```

```python
total_bill_thur[total_bill_thur > (Q3 + 1.5 * IQR)]
```

```python
g = sns.jointplot("total_bill", "tip", data=tips, kind="reg")
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1);
```
